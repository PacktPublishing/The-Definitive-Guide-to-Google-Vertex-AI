#### In this notebook we analyse the correlation of features with the target variable 'deposit_signup''

```python
import pandas as pd
import numpy as np
```

```python
# Load data
data = pd.read_csv('./dataset/bank-subset.csv')
```

```python
# Convert categorical variable to dummy variables
data_encoded = pd.get_dummies(data)

# Compute correlation matrix
corr = data_encoded.corr()
```

```python
# Display correlation matrix
corr['deposit_signup_yes'].sort_values(ascending=False)
```
